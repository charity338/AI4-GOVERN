## Create Contract Value Percentile

### Subtask:
Calculate and add a new column for the percentile of each 'Supplier Contract Amount (USD)' to understand the contract size distribution.


**Reasoning**:
To calculate the percentile of each 'Supplier Contract Amount (USD)' and add it as a new column, I will use the `rank(pct=True)` method and then multiply by 100 to get percentages. Finally, I will display the first few rows of the DataFrame to verify the new column.



```python
df_filtered['Contract Value Percentile'] = df_filtered['Supplier Contract Amount (USD)'].rank(pct=True) * 100
print("First 5 rows of df_filtered with 'Contract Value Percentile':")
print(df_filtered.head())
```

**Reasoning**:
To explicitly avoid the `SettingWithCopyWarning`, I will use `.loc` when creating the new column in `df_filtered`, ensuring the operation is performed on the DataFrame directly.



```python
df_filtered.loc[:, 'Contract Value Percentile'] = df_filtered['Supplier Contract Amount (USD)'].rank(pct=True) * 100
print("First 5 rows of df_filtered with 'Contract Value Percentile':")
print(df_filtered.head())
```

**Reasoning**:
To identify repeat suppliers, I will first count the occurrences of each supplier in the 'Supplier' column. Then, I will create a new boolean column 'Repeat Supplier Flag' in the `df_filtered` DataFrame, setting it to `True` for suppliers with more than one contract and `False` otherwise. Finally, I will display the first few rows and the value counts of the new column to verify the change.



```python
supplier_counts = df_filtered['Supplier'].value_counts()
df_filtered.loc[:, 'Repeat Supplier Flag'] = df_filtered['Supplier'].map(supplier_counts) > 1

print("First 5 rows of df_filtered with 'Repeat Supplier Flag':")
print(df_filtered.head())
print("\nValue counts for 'Repeat Supplier Flag':")
print(df_filtered['Repeat Supplier Flag'].value_counts())
```

**Reasoning**:
To extract the contract signing year, I will first convert the 'Contract Signing Date' column to datetime objects using `pd.to_datetime` and then extract the year using the `.dt.year` accessor. Finally, I will display the first few rows of the DataFrame to verify the new column.



```python
df_filtered.loc[:, 'Contract Signing Year'] = pd.to_datetime(df_filtered['Contract Signing Date']).dt.year

print("First 5 rows of df_filtered with 'Contract Signing Year':")
print(df_filtered.head())
```

**Reasoning**:
To calculate the count of contracts per 'Borrower Country / Economy' and 'Project Global Practice', I will group the DataFrame by these columns separately and use `transform('count')` to get the counts for each row. This will create new columns with the respective contract counts. Finally, I will display the first few rows of the DataFrame to verify these new columns.



```python
df_filtered.loc[:, 'Contracts per Borrower Country'] = df_filtered.groupby('Borrower Country / Economy')['Project ID'].transform('count')
df_filtered.loc[:, 'Contracts per Project Global Practice'] = df_filtered.groupby('Project Global Practice')['Project ID'].transform('count')

print("First 5 rows of df_filtered with new contract count columns:")
print(df_filtered.head())
```

**Reasoning**:
To create the 'Contract Risk Labels' column, I will define risk categories based on 'Contract Value Percentile' and 'Repeat Supplier Flag'. I will use numpy's `select` function to apply conditional logic to assign 'High', 'Medium', or 'Low' risk labels, and then display the value counts and first few rows to verify the new column.



```python
import numpy as np

conditions = [
    (df_filtered['Contract Value Percentile'] >= 75) & (df_filtered['Repeat Supplier Flag'] == False),
    (df_filtered['Contract Value Percentile'] < 50) & (df_filtered['Repeat Supplier Flag'] == True)
]
choices = ['High', 'Low']

df_filtered.loc[:, 'Contract Risk Labels'] = np.select(conditions, choices, default='Medium')

print("First 5 rows of df_filtered with 'Contract Risk Labels':")
print(df_filtered.head())
print("\nValue counts for 'Contract Risk Labels':")
print(df_filtered['Contract Risk Labels'].value_counts())
```

## Summary of New Features and Impact on Risk Analysis

### New Features Added:
1.  **Contract Value Percentile**: This column provides a normalized view of each contract's value relative to all other contracts. For example, a contract with a percentile of 90 means its value is greater than 90% of all other contracts.
2.  **Repeat Supplier Flag**: A boolean flag (`True`/`False`) indicating whether a supplier has previously been awarded contracts. This helps distinguish new suppliers from those with established relationships.
3.  **Contract Signing Year**: The year when each contract was signed, extracted from the 'Contract Signing Date'. This allows for time-series analysis and understanding contract activity over different years.
4.  **Contracts per Entity/Practice**: Columns for the count of contracts per 'Borrower Country / Economy' and 'Project Global Practice'.
5.  **Contracts per Borrower Country**: The total count of contracts associated with each 'Borrower Country / Economy'. This highlights which countries are most active in terms of contract volume.
6.  **Contracts per Project Global Practice**: The total count of contracts within each 'Project Global Practice'. This shows which sectors or global practices have the highest contract activity.
7.  **Contract Risk Labels**: A categorical label ('High', 'Medium', 'Low') assigned based on a combination of 'Contract Value Percentile' and 'Repeat Supplier Flag'.
    *   **High Risk**: Contracts with a 'Contract Value Percentile' >= 75 and 'Repeat Supplier Flag' == False (large, non-repeat supplier).
    *   **Low Risk**: Contracts with a 'Contract Value Percentile' < 50 and 'Repeat Supplier Flag' == True (smaller, repeat supplier).
    *   **Medium Risk**: All other contracts.

### Potential Impact on Risk Analysis:
*   **Contract Value Percentile**: Helps identify unusually large contracts that might warrant extra scrutiny due to higher financial exposure. Outliers in this feature could indicate potential areas of concern.
*   **Repeat Supplier Flag**: Repeat suppliers might imply a lower risk due to established trust and performance history, or conversely, might indicate a lack of competition or potential for collusion if not properly managed. New suppliers might represent higher risk if their capabilities are unproven.
*   **Contract Signing Year**: Allows for trend analysis of contract awards, identifying periods of increased or decreased activity. Unusual spikes or drops in specific years could signal changes in procurement strategy or external factors affecting project execution.
*   **Contracts per Borrower Country** and **Contracts per Project Global Practice**: These features can reveal concentration risks. If a large number of contracts are concentrated in a single country or a specific global practice, it might indicate over-reliance or vulnerability to localized issues. Conversely, a low number might suggest less mature markets or less significant projects, potentially impacting risk profiles.
*   **Contract Risk Labels**: This composite feature provides a quick, high-level categorization of contracts based on predefined risk criteria. It can be used to prioritize contracts for deeper due diligence, allocate monitoring resources more effectively, and highlight areas needing policy adjustments. For instance, a high proportion of 'High' risk contracts might signal a need to review procurement guidelines or supplier vetting processes.
